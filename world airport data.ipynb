```python
import pandas as pd
data = pd.read_csv('airports.csv')
data
```

```python
import matplotlib.pyplot as plt
```

```python
data.head(1)
```

```python
%matplotlib inline
```

```python
#Can you create a bar chart to visualize the top 10 countries with the highest number of airports?
country_airports = data['iso_country'].value_counts().head(10)
plt.bar(country_airports.index, country_airports.values)
plt.xlabel('countries')
plt.ylabel('number of airports')
plt.title('top 10 coutnries witht he highest number of airports')
plt.xticks(rotation=45)
plt.show()
```

```python
#showing the different types of airports with it's respective elevation feet through box plot.What is the average number of runways per airport? Are there any outliers?

import seaborn as sns

sns.boxplot(x='type',y='elevation_ft', data=data)
plt.xlabel('Type')
plt.ylabel('Elevation_ft')
plt.title('Distribution of different types of airports and its elevation feet')
plt.xticks(rotation=90)
plt.show()
```

```python
#Can you create a scatter plot to show the relationship between the latitude and longitude of airports?

plt.scatter(data['longitude_deg'],data['latitude_deg'])
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('scatter plot of airport locations')
plt.show()
```
