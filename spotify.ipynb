```python
#for mathematical calculations
import numpy as np
import pandas as pd
import scipy.stats as stats

#for data visualization
import seaborn as sns
import matplotlib.pyplot as plt
import plotly
import plotly.express as px
from matplotlib.pyplot import figure
```

```python
df = pd.read_csv("/spotify_dataset.csv", encoding='latin-1')
df.head()
```

```python
df.info()
```

```python
#number of times charted by artist

df_numbercharted = df.groupby('Artist').sum().sort_values('Number of Times Charted', ascending=False)
df_numbercharted=df_numbercharted.reset_index()
df_numbercharted
```

```python
#graph

px.bar(x='Artist', y='Number of Times Charted', data_frame=df_numbercharted.head(7), title="Top 7 Artist with Highest Number of Charted")
```

```python
#correlation
#first step to clean data
```

```python
df=df.fillna('')
df=df.replace(' ','')
df['Streams']=df['Streams'].str.replace(',','')
```

```python
 #convert columns to numeric
df[['Highest Charting Position', 'Number of Times Charted', 'Streams','Popularity','Danceability','Energy','Loudness','Speechiness','Acousticness','Liveness','Tempo','Duration (ms)','Valence']]=df[['Highest Charting Position', 'Number of Times Charted', 'Streams','Popularity','Danceability','Energy','Loudness','Speechiness','Acousticness','Liveness','Tempo','Duration (ms)','Valence']].apply(pd.to_numeric)
```

```python
#separate year from the column "Release Date"to be able to analyze the correlations
```

```python
df['Release Year']=pd.DatetimeIndex(df['Release Date']).year
```

```python
%matplotlib inline
f,ax=plt.subplots(figsize=(14,10))
sns.heatmap(df.corr(),annot=True, fmt=".1f",ax=ax)
plt.show()
```

```python
#danceability

px.line(x='Release Year', y='Danceability', data_frame=df, title="Danceability over the course of the year" )
```

```python
#number of times charted correlation with years

dfyear=df.groupby('Release Year').sum().sort_values('Number of Times Charted', ascending=False)
dfyear=dfyear.reset_index()

px.bar(x='Release Year',y='Number of Times Charted',data_frame=dfyear.head(7))
```

```python
artistbypop=df.groupby('Artist').sum().sort_values('Popularity', ascending=False)[:20]
artistbypop=artistbypop.reset_index()
px.bar(x='Artist',y='Popularity', data_frame=artistbypop)
```

```python
#most popular genres
df['Genre']=df["Genre"].astype(str)
df["Genre"][df["Genre"] == "[]" ]=np.nan
df['Genre']=df["Genre"].fillna(0)

#we get rid of unwanted symbols to be able to separate genres
df.Genre=df.Genre.str.replace("[","")
df.Genre=df.Genre.str.replace("]","")
df.Genre=df.Genre.str.replace("'","")

#we divide genre strings by comma
df['Genre']=df["Genre"].str.split(",")

#separate rows based on genres, so for each song that is marked with several genres will have multiple rows with one genre
df=df.explode('Genre')
df
```

```python
 fig = plt.figure(figsize=(10,10))
 ax=fig.subplots()
 df.Genre.value_counts()[:30].plot(ax=ax, kind='pie')
 ax.set_ylabel("")
 ax.set_title("Top 30 most popular genres")
 plt.show()
```
